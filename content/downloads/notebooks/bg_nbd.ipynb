```python
import os

import numpy as np
import pymc as pm
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DATA_DIR = os.path.join(os.getcwd(), 'data/')
CHART_DIR = os.path.join(os.getcwd(), 'charts/')
```

"The Pareto/NBD (and, as we will see shortly, the BG/NBD) requires only two pieces of information about each customer’s past purchasing history: his “recency” (when his last trans- action occurred) and “frequency” (how many transactions he made in a specified time period). The notation used to represent this information is (X = x,tx,T), where x is the number of transactions observed in the time period (0,T] and tx (0 < tx ≤ T) is the time of the last transaction."

```python
data_file = DATA_DIR + 'cdnow_data.csv'
df = pd.read_csv(data_file, index_col='ID')
df['made_a_transaction'] = (df.x > 0).astype(int)
df.head()
```

```python
num_users = df.shape[0]
transactions_in_period = df.x.values
made_a_transaction_dummy = df.made_a_transaction.values
time_of_last_transaction = df.t_x.values
time_period_end = df.T.values
as_array = df.values
```

```python
df['starting_trans_rates']= df.x / df['T']
df.starting_trans_rates = df.starting_trans_rates.apply(lambda x: max(x, 0.05))
```

@pm.stochastic
def transactions(x=transactions_in_period, 
                 x_gt_0=made_a_transaction_dummy, 
                 t_x=time_of_last_transaction, 
                 T=time_period_end, 
                 lambdas=transaction_rates, 
                 p=dropout_rates, value=-np.inf):
    return np.sum(-np.log(((1-p)**x)*(lambdas**x)*(np.exp(-1*lambdas*T)) + \
            x_gt_0*p*((1-p)**(x-1))*(lambdas**x)*(np.exp(-1*lambdas*t_x))))

```python
#hyperpriors
gamma_a = pm.Uniform('gamma_a', 0.00001, 100, value=0.243)
gamma_b = pm.Uniform('gamma_b', 0.00001, 10, value=4.414)
beta_a = pm.Uniform('beta_a', 0.00001, 1000, value=.79)
beta_b = pm.Uniform('beta_b', 0.00001, 1000, value=2.46)

#customer-specific parameters
transaction_rates = pm.Gamma('transaction_rates', 
                               alpha=gamma_a,
                               beta=gamma_b,
                               size=num_users, value=df.starting_trans_rates.values)

dropout_rates = pm.Beta("dropout_rates", 
                        alpha=beta_a,
                        beta=beta_b,
                        size=num_users)
@pm.observed
def transactions(lambdas=transaction_rates, p=dropout_rates, value=as_array):
    def logp(value, lambdas, p):
        x=value[:,0]
        t_x=value[:,1]
        T=value[:,2]
        x_gt_0=value[:,3]
        return np.sum(np.log(((1-p)**x)*(lambdas**x)*(np.exp(-1*lambdas*T)) + \
                x_gt_0*p*((1-p)**(x-1))*(lambdas**x)*(np.exp(-1*lambdas*t_x))))

mcmc = pm.MCMC([gamma_a, gamma_b, beta_a, beta_b, transaction_rates, dropout_rates, transactions])
#map_ = pm.MAP( mcmc )
#map_.fit()
```
