## Clearly, we need do some munging here.  Let's turn this into a long dataframe, and split the score into two numeric columns.

```python
def clean_wiki_score_table(filename):    
    df = pd.read_csv(filename, sep='\t', index_col=0,)
    df.index = df.columns
    rows = []
    for i in df.index:
        for c in df.columns:
            if i == c: continue
            score = df.ix[i, c]
            if pd.isnull(score) or score == 'a': continue
            score = [int(row) for row in score.split('-')]
            rows.append([i, c, score[0], score[1]])
    df = pd.DataFrame(rows, columns = ['home', 'away', 'home_score', 'away_score'])
    
    teams = df.home.unique()
    teams = pd.DataFrame(teams, columns=['team'])
    teams['i'] = teams.index
    
    df = pd.merge(df, teams, left_on='home', right_on='team', how='left')
    df = df.rename(columns = {'i': 'i_home'}).drop('team', 1)
    df = pd.merge(df, teams, left_on='away', right_on='team', how='left')
    df = df.rename(columns = {'i': 'i_away'}).drop('team', 1)
    
    return df, teams

df_2013, teams_2013 = clean_wiki_score_table(data_dir + 'premier_league_13_14.txt')
df_2014, teams_2014 = clean_wiki_score_table(data_dir + 'premier_league_14_15.txt')
```

```python
observed_home_goals_2013 = df_2013.home_score.values
observed_away_goals_2013 = df_2013.away_score.values

observed_home_goals_2014 = df_2014.home_score.values
observed_away_goals_2014 = df_2014.away_score.values

home_team_2013 = df_2013.i_home.values
away_team_2013 = df_2013.i_away.values

home_team_2014 = df_2014.i_home.values
away_team_2014 = df_2014.i_away.values

num_teams_2013 = len(df_2013.i_home.unique())
num_teams_2014 = len(df_2014.i_home.unique())
assert num_teams_2013 == num_teams_2014
```

```python
g = df_2013.groupby('i_away')
att_starting_points = np.log(g.away_score.mean())
g = df_2013.groupby('i_home')
def_starting_points = -np.log(g.away_score.mean())
```

Now let's define the pymc model, mimicing the model used in the paper.

```python
#hyper parameters
home = pymc.Normal('home', 0, .0001, value=0)
#tau_att = pymc.Gamma('tau_att', .1, .1, value=10)
#tau_def = pymc.Gamma('tau_def', .1, .1, value=10)
#tau_seasonal = pymc.Gamma('tau_seasonal', .1, .1, value=10)
tau_att = pymc.Uniform('tau_att', 0, 100, value=10)
tau_def = pymc.Uniform('tau_def', 0, 100, value=10)
tau_seasonal = pymc.Uniform('tau_seasonal', 0, 100, value=10)
intercept = pymc.Normal('intercept', 0, .0001, value=0)

#team-specific parameters
atts_star_2013 = pymc.Normal("atts_star_2013", mu=0, tau=tau_att, size=num_teams_2013, value=att_starting_points)
defs_star_2013 = pymc.Normal("defs_star_2013", mu=0, tau=tau_def, size=num_teams_2013, value=def_starting_points)

atts_star_2014 = np.empty(num_teams_2014, dtype=object)
defs_star_2014 = np.empty(num_teams_2014, dtype=object)

for i in range(num_teams_2014):
    slug = teams_2014.team[i]
    if slug in teams_2013.team.unique():
        i_2013 = teams_2013.i[teams_2013.team == slug]    
        atts_star_2014[i] = pymc.Normal("atts_2014_%i" % i, 
                                        mu=atts_star_2013[i_2013], 
                                        tau=tau_seasonal, 
                                        value=atts_star_2013[i_2013].value)
        defs_star_2014[i] = pymc.Normal("defs_2014_%i" % i, 
                                        mu=defs_star_2013[i_2013], 
                                        tau=tau_seasonal, 
                                        value=defs_star_2013[i_2013].value)
    else:
        atts_star_2014[i] = pymc.Normal("atts_2014_%i" % i, mu=0, tau=tau_att, value=0)
        defs_star_2014[i] = pymc.Normal("defs_2014_%i" % i, mu=0, tau=tau_att, value=0)
        

# trick to code the sum to zero contraint
@pymc.deterministic
def atts_2013(atts_star_2013=atts_star_2013):
    atts = atts_star_2013.copy()
    #return atts - np.mean(atts_star_2013)
    return atts

@pymc.deterministic
def defs_2013(defs_star_2013=defs_star_2013):
    defs = defs_star_2013.copy()
    #return defs - np.mean(defs_star_2013)
    return defs

@pymc.deterministic
def atts_2014(atts_star_2014=atts_star_2014):
    atts = [_ for _ in atts_star_2014] # prevent segfault 11
    #return atts - np.mean(atts_star_2014)
    return atts

@pymc.deterministic
def defs_2014(defs_star_2014=defs_star_2014):
    defs = [_ for _ in defs_star_2014] # prevent segfault 11
    #return defs - np.mean(defs_star_2014)
    return defs
    
@pymc.deterministic
def home_theta_2013(home_team_2013=home_team_2013, away_team_2013=away_team_2013, home=home, 
                    intercept=intercept, atts_2013=atts_2013, defs_2013=defs_2013):
    return np.exp(intercept + home + atts_2013[home_team_2013] + defs_2013[away_team_2013])

@pymc.deterministic
def away_theta_2013(home_team_2013=home_team_2013, away_team_2013=away_team_2013, home=home, 
                    intercept=intercept, atts_2013=atts_2013, defs_2013=defs_2013):
    return np.exp(intercept + atts_2013[away_team_2013] + defs_2013[home_team_2013])

@pymc.deterministic
def home_theta_2014(home_team_2014=home_team_2014, away_team_2014=away_team_2014, home=home, 
                    intercept=intercept, atts_2014=atts_2014, defs_2014=defs_2014):
    return np.exp(intercept + home + atts_2014[home_team_2014] + defs_2014[away_team_2014])

@pymc.deterministic
def away_theta_2014(home_team_2014=home_team_2014, away_team_2014=away_team_2014, home=home, 
                    intercept=intercept, atts_2014=atts_2014, defs_2014=defs_2014):
    return np.exp(intercept + atts_2014[away_team_2014] + defs_2014[home_team_2014])

home_goals_2013 = pymc.Poisson('home_goals_2013', mu=home_theta_2013, value=observed_home_goals_2013, observed=True)
away_goals_2013 = pymc.Poisson('away_goals_2013', mu=away_theta_2013, value=observed_away_goals_2013, observed=True)

home_goals_2014 = pymc.Poisson('home_goals_2014', mu=home_theta_2014, value=observed_home_goals_2014, observed=True)
away_goals_2014 = pymc.Poisson('away_goals_2014', mu=away_theta_2014, value=observed_away_goals_2014, observed=True)

@pymc.potential 
def ensure_taus_in_control(tau_att=tau_att, tau_def=tau_def, tau_seasonal=tau_seasonal):
    """
    necessary because pymc doesn't respect bounds on uniform
    """
    if tau_att > 100 or tau_def > 100:
        return -1e6
    if tau_att < 5 or tau_def < 5 or tau_seasonal < 5:
        return -1e6
    return 0 


mcmc = pymc.MCMC([home, intercept, tau_att, tau_def, tau_seasonal,
                  atts_star_2013, atts_star_2014, defs_star_2013, defs_star_2014,
                  atts_2013, defs_2013, atts_2014, defs_2014, 
                  home_theta_2013, home_theta_2014, away_theta_2013, away_theta_2014,
                  home_goals_2013, away_goals_2013, home_goals_2014, away_goals_2014])
#map_ = pymc.MAP( mcmc )
#map_.fit()
mcmc.sample(100000, 10000, 8)
```

### Simulate a season

```python
def simulate_season():
    num_samples = atts.trace().shape[0]
    draw = np.random.randint(0, num_samples)
    atts_draw = pd.DataFrame({'att': atts.trace()[draw, :],})
    defs_draw = pd.DataFrame({'def': defs.trace()[draw, :],})
    home_draw = home.trace()[draw]
    intercept_draw = intercept.trace()[draw]
    season = df.copy()
    season = pd.merge(season, atts_draw, left_on='i_home', right_index=True)
    season = pd.merge(season, defs_draw, left_on='i_home', right_index=True)
    season = season.rename(columns = {'att': 'att_home', 'def': 'def_home'})
    season = pd.merge(season, atts_draw, left_on='i_away', right_index=True)
    season = pd.merge(season, defs_draw, left_on='i_away', right_index=True)
    season = season.rename(columns = {'att': 'att_away', 'def': 'def_away'})
    season['home'] = home_draw
    season['intercept'] = intercept_draw
    season['home_theta'] = season.apply(lambda x: math.exp(x['intercept'] + 
                                                           x['home'] + 
                                                           x['att_home'] + 
                                                           x['def_away']), axis=1)
    season['away_theta'] = season.apply(lambda x: math.exp(x['intercept'] + 
                                                           x['att_away'] + 
                                                           x['def_home']), axis=1)
    season['home_goals'] = season.apply(lambda x: np.random.poisson(x['home_theta']), axis=1)
    season['away_goals'] = season.apply(lambda x: np.random.poisson(x['away_theta']), axis=1)
    season['home_outcome'] = season.apply(lambda x: 'win' if x['home_goals'] > x['away_goals'] else 
                                                    'loss' if x['home_goals'] < x['away_goals'] else 'draw', axis=1)
    season['away_outcome'] = season.apply(lambda x: 'win' if x['home_goals'] < x['away_goals'] else 
                                                    'loss' if x['home_goals'] > x['away_goals'] else 'draw', axis=1)
    season = season.join(pd.get_dummies(season.home_outcome, prefix='home'))
    season = season.join(pd.get_dummies(season.away_outcome, prefix='away'))
    return season


def create_season_table(season):
    g = season.groupby('i_home')    
    home = pd.DataFrame({'home_goals': g.home_goals.sum(),
                         'home_goals_against': g.away_goals.sum(),
                         'home_wins': g.home_win.sum(),
                         'home_draws': g.home_draw.sum(),
                         'home_losses': g.home_loss.sum()
                         })
    g = season.groupby('i_away')    
    away = pd.DataFrame({'away_goals': g.away_goals.sum(),
                         'away_goals_against': g.home_goals.sum(),
                         'away_wins': g.away_win.sum(),
                         'away_draws': g.away_draw.sum(),
                         'away_losses': g.away_loss.sum()
                         })
    df = home.join(away)
    df['wins'] = df.home_wins + df.away_wins
    df['draws'] = df.home_draws + df.away_draws
    df['losses'] = df.home_losses + df.away_losses
    df['points'] = df.wins * 3 + df.draws
    df['gf'] = df.home_goals + df.away_goals
    df['ga'] = df.home_goals_against + df.away_goals_against
    df['gd'] = df.gf - df.ga
    df = pd.merge(teams, df, left_on='i', right_index=True)
    df = df.sort_index(by='points', ascending=False)
    return df  
    
def simulate_seasons(n=100):
    dfs = []
    for i in range(n):
        s = simulate_season()
        t = create_season_table(s)
        t['iteration'] = i
        dfs.append(t)
    return pd.concat(dfs, ignore_index=True)
    
```
